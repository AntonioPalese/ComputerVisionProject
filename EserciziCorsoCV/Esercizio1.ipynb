```python
import torch
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Reading and displaying a grayscale image
path = './Images/Lil-Peep-HD-Desktop-Background.jpg'
im = cv2.imread(path, cv2.IMREAD_GRAYSCALE) # BGR, (H, W)
plt.imshow(im, cmap='gray')
plt.show()

```

```python
im = cv2.imread(path,cv2.IMREAD_COLOR).swapaxes(0,2) # BGR, (3, H, W)
im.shape
im.swapaxes(0,2).shape
```

```python
plt.imshow(im[::-1].swapaxes(0,2)) # Reconvert to RGB and to (H, W, 3) for visualization
```

```python
cv2.imshow("Vision", im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

'''Your code will take as input a color image (a torch.Tensor with dtype torch.uint8 and rank 3) and two
scalars a and b. It must apply a pixel-wise linear transformation (every pixel p is transformed to a*p +
b). The code should produce a new image with the same shape and dtype. a and b can be either ints
or floats.
 The scale factor a is also called gain or contrast
 The offset b is also called bias or brightness.
 Be careful to: compute the exact result (in torch.float32), round to nearest integer and then clip
between 0 and 255.'''

a = 1
b = 92

img = np.clip(np.round(im*a + b), 0, 255).astype(np.uint8)
img.to
```

```python
plt.imshow(img[::-1].swapaxes(0,2)) # Reconvert to RGB and to (H, W, 3) for visualization
```
