```python
from io import BytesIO
import numpy as np
import cv2
import torch
from skimage import data
import matplotlib.pyplot as plt

im = torch.from_numpy(data.coins()[160:230, 70:270])
```

```python
plt.imshow(im)
```

```python
edges = cv2.Canny(im.numpy(),200,800)
```

```python
plt.imshow(edges)
```

```python
rows = im.shape[0]
circles = cv2.HoughCircles(edges, cv2.HOUGH_GRADIENT, 1, minDist = 40,
                               param1=40, param2=20,
                               minRadius=10, maxRadius=0)
circles.shape
```

```python
detected_circles = np.uint16(np.around(circles))
detected_circles.shape
```

```python
i = 0
output = im.numpy().copy()
for (x, y, r) in detected_circles[0, :]:
    cv2.circle(output, (x, y), r, (0, 100, 255), 2)
    cv2.circle(output, (x, y), 1, (0, 100, 255), 2)
    print(0)
plt.imshow(output)
```
