```python
# loop over the input images
#path = 'BlurredImages/blurred04.jpg'

import argparse
import cv2
import matplotlib.pyplot as plt
import numpy as np


def variance_of_laplacian(image):
    # compute the Laplacian of the image and then return the focus
    # measure, which is simply the variance of the Laplacian
    return cv2.Laplacian(image, cv2.CV_64F)





path1 = 'BlurredImages/blurred01.jpg'
path2 = 'BlurredImages/blurred02.jpg'

image1 = cv2.imread(path1)
image2 = cv2.imread(path2)[:183, :275, :]

image1 = image1.reshape((1, 183, 275, 3))
image2 = image2.reshape((1, 183, 275, 3))

image = np.concatenate((image1, image2), axis = (0))

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
imr = variance_of_laplacian(gray)
fm = imr.var()
print(imr.shape)
text = "Not Blurry"
    
if fm < 134:
    text = "Blurry"
# show the image
cv2.putText(image, "{}: {:.2f}".format(text, fm), (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 3)
plt.imshow(imr, cmap='gray')
plt.show()
cv2.imshow("Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
